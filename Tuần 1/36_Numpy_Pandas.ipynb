```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
a = np.array([1, 2, 3, 4])
b = np.zeros((3, 4))
c = np.ones(5)
d = np.arange(0, 10, 2)
e = np.linspace(0, 1, 5)
f = np.random.rand(3, 3)
g = np.random.randn(2, 3)

print('a shape:', a.shape, ' dtype:', a.dtype)
print('b =\n', b)
print('e =', e)
print('f =\n', f)
print('g =\n', g)
```

```python
#Reshape, indexing, slicing
x = np.arange(12)            # 0, 1, ..., 11
X = x.reshape(3, 4)
print('X =\n', X)

# Lấy phần tử (giống Python)
print('X[0, 0] =', X[0, 0])# góc trên trái
print('X[1, 2] =', X[1, 2])# hàng 1, cột 2

# Slicing: hàng 1, tất cả cột
print('X[1, :] =', X[1, :])

# Cột cuối cùng
print('X[:, -1] =', X[:, -1])

# Boolean indexing — cực kỳ hữu ích
mask = X > 5
print('mask =\n', mask)
print('X[X > 5] =', X[X > 5])
```

```python
#Broadcasting — phép màu của NumPy
M = np.arange(12).reshape(3,4)
v = np.array([10, 20, 30, 40])
print('M + v =\n', M + v)
M_centered = M - M.mean(axis=0)
print('M_centered =\n', M_centered)
print('Mean theo cột:', M_centered.mean(axis=0))
```

```python
# Phép nhân ma trận và dot product
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print('A * B (element-wise) =\n', A * B)
print('A @ B (matrix mul)    =\n', A @ B)

# Dot product cho vector
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])
print('u . v =', u @ v)
```

```python
#Axis — chỗ hay nhầm nhất
X = np.arange(12).reshape(3, 4)
print('X =\n', X)
print('Tổng tất cả        :', X.sum())
print('Tổng theo axis=0 (gộp hàng → 1 hàng):', X.sum(axis=0))
print('Tổng theo axis=1 (gộp cột → 1 cột):', X.sum(axis=1))
print('Argmax theo axis=1 :', X.argmax(axis=1))
```

```python
df = pd.DataFrame({
    'name': ['An', 'Binh', 'Cuong', 'Dung', 'Em'],
    'age':  [20, 22, 21, np.nan, 23],
    'score': [8.5, 7.0, 9.0, 6.5, 8.0],
    'major': ['CNTT', 'CNTT', 'AI', 'AI', 'CNTT'],
})
df
```

```python
print(df.shape)
print(df.columns.tolist())
print(df.dtypes)
print()
df.describe()
```

```python
print('Số NaN trong mỗi cột:')
print(df.isna().sum())
print()
print('Bỏ hàng có NaN:')
print(df.dropna())
print()
print('Điền NaN bằng giá trị trung bình của cột:')
print(df.fillna({'age': df['age'].mean()}))
```

```python
# Trung bình điểm theo ngành
print(df.groupby('major')['score'].mean())
print()

# Nhiều phép gộp một lúc
print(df.groupby('major').agg(
    so_sinh_vien=('name', 'count'),
    diem_tb=('score', 'mean'),
    diem_max=('score', 'max'),
))
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hàm sin và cos')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
#Scatter với nhiều lớp màu
x1 = np.random.randn(100,2) + np.array([2, 2])
x2 = np.random.randn(100,2) + np.array([-2, -2])

fig,axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].scatter(x1[:, 0], x1[:, 1], color='red', label='Lớp A', alpha=0.6)
axes[0].scatter(x2[:, 0], x2[:, 1], color='blue', label='Lớp B',alpha=0.6)
axes[0].legend()
axes[0].set_title('Hai cụm điểm');axes[0].grid(alpha=0.3)

axes[1].hist(np.random.randn(2000),bins=30,color='yellow',alpha=0.7)
axes[1].set_title('Phân phối chuẩn N(0, 1) — 2000 mẫu')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
from google.colab import files
import io
print("Vui lòng chọn file Iris.csv để upload:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
iris = pd.read_csv(io.BytesIO(uploaded[file_name]))
display(iris.head())
```

```python
print(f"Số dòng: {iris.shape[0]}, Số cột: {iris.shape[1]}")
print("\nTrung bình sepal.length theo từng loài:")
display(iris.groupby('variety')['sepal.length'].mean())
max_petal_width_row = iris.loc[iris['petal.width'].idxmax()]
print(f"\nLoài có petal.width lớn nhất là: {max_petal_width_row['variety']} ({max_petal_width_row['petal.width']})")
```

```python
plt.figure(figsize=(10, 6))
varieties = iris['variety'].unique()
colors = ['red', 'green', 'blue']

for variety, color in zip(varieties, colors):
    subset = iris[iris['variety'] == variety]
    plt.scatter(subset['sepal.length'], subset['petal.length'],
                label=variety, color=color, alpha=0.7)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris: Sepal Length vs Petal Length')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
